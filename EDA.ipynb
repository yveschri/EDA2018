### 5°Backers

```python
plt.figure(figsize=(20,12))
sns.barplot(x='main_category',y='backers',data=Tableau)
```

#### On obverse que le nombre de backers est élevé dans Design, Games, Technology et Comic, pourtant le nombre de réussite ou le taux de réussite n'est pas élevé dans ces catégories,  cela veut dire qu'un grand nombre de backers n'assurent pas forcément la réussite d'un projet

```python
Tableau['avg_pledged'] = Tableau['usd_pledged_real'] / Tableau['backers']

Tableau['avg_pledged'].fillna(0, inplace=True)
Tableau.replace(to_replace=[np.inf, -np.inf], value=0, inplace=True)


cat_list = list(Tableau['main_category'].unique())


pledge_mapping = dict()
for category in cat_list:
    cat_Tableau = Tableau[Tableau['main_category'] == category]
    pledge_mapping.update({category : cat_Tableau['avg_pledged'].mean()})
    
pledge_Tableau= pd.DataFrame(pledge_mapping, index=['avg_pledged'])
pledge_Tableau.sort_values('avg_pledged', axis=1, inplace=True, ascending=False)
display(pledge_Tableau.head())
    

    
```

#### On observe que les backers en moyenne donnent beaucoup plus en Technology et en Design que en Games ou Comics, pourtant ces derniers ont un taux de réussite supérieur au deux premiers
